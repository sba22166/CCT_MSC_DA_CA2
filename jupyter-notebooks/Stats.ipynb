# Confidence Intervals

Milk Prices

```python
def calculate_ci(data, confidence_level=0.95):
    mean, margin_of_error = scipy.stats.t.interval(confidence_level, len(data) - 1, loc=np.mean(data), scale=scipy.stats.sem(data))
    return mean, margin_of_error

data_ireland = df_ireland["MilkPrice (Per 100 KG €)"]
mean_ireland, margin_of_error_ireland = calculate_ci(data_ireland)

data_germany = df_germany["MilkPrice (Per 100 KG €)"]
mean_germany, margin_of_error_germany = calculate_ci(data_germany)

data_spain = df_spain["MilkPrice (Per 100 KG €)"]
mean_spain, margin_of_error_spain = calculate_ci(data_spain)

print(f"CI of Irish Milk Prices: [{mean_ireland}, {margin_of_error_ireland}]")
print(f"CI of German Milk Prices: [{mean_germany}, {margin_of_error_germany}]")
print(f"CI of Spain Milk Prices: [{mean_spain}, {margin_of_error_spain}]")
```

# T Test

Becuase the Price of Dairy milk in Ireland and Germany is simialr , which we learned above in the results of the Confidence Intervals, lets compare data from Ireland and Germany for these tests first.


```python
#Creating test data for the milk value for tesing.

test1 = df_ireland["MilkPrice (Per 100 KG €)"]
test2 = df_germany["MilkPrice (Per 100 KG €)"]

stat, p_value = ttest_ind(test1, test2) 
print("Ireland V Germany - Milk Prices - Statistics =%.3f, p_value=%.3f" % (stat, p_value))

# Equates to our 95% confidence interval
alpha = 0.05
# simple if/else statment to print out whether we reject the (H0) or cannot reject it.
if p_value > alpha:
    print("p_value is greater than 0.05, so we cannot  reject the null hypothesis(H0).")
else:
    print("p_value is less than or equal to 0.05, so we can reject the null hypothesis(H0).")
```



Lets also test Ireland's data V Spain for Milk compared to ireland and compare.



```python
test3 = df_spain["MilkPrice (Per 100 KG €)"]

stat, p_value = ttest_ind(test1, test3) 
print("Ireland V Spain - Milk Prices - Statistics =%.3f, p_value=%.3f" % (stat, p_value))

# Equates to our 95% confidence interval
alpha = 0.05
# simple if/else statment to print out whether we reject the (H0) or cannot reject it.
if p_value > alpha:
    print("p_value is greater than 0.05, so we cannot  reject the null hypothesis(H0).")
else:
    print("p_value is less than or equal to 0.05, so we can reject the null hypothesis(H0).")
```
